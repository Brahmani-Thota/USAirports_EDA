The data provided is in .tsv format and the space divided didnt have tab space(i.e., '\t'). It had fixed space format to show the data uniform. This was one of the first problem.

Links Used:https://python-forum.io/Thread-Is-there-any-way-to-properly-load-fixed-width-file-into-a-dataframe-using-Pandas

```python
airports=pd.read_fwf('sample.tsv', widths=[20,23,7,6,9,10,10])
airports
```

After reading the data, columns were added.

Links used:https://datascience.stackexchange.com/questions/45314/dataframe-has-no-column-names-how-to-add-a-header

```python
airports.columns=['Airport','City','Scheduled departures','Performed departures','Enplaned passengers','Enplaned revenue tons of freight','Enplaned revenue tons of mail']
airports
```

```python
#Checking for missing values

airports.isnull().sum()
```

```python
#To check for the number of airports in each city

airports['City'].value_counts()
```

```python
#To check for passengers in each city 

airports.groupby('City')['Enplaned passengers'].first()
```

```python
#To check for the departures in each city

#Links used:https://www.geeksforgeeks.org/python-pandas-dataframe-groupby/
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.core.groupby.GroupBy.mean.html

airports.groupby('City')['Performed departures'].mean()
```

```python
#The number of cities present
airports['City'].nunique()
```

Scatter Plot
Reason: To know the distribution of departures of various cities

```python
_=airports.plot(x='City', y='Scheduled departures', style='*')
plt.ylabel('Count of Scheduled departures')
plt.xlabel('Cities')
plt.xlim(0,100)
plt.xticks(rotation=90)
plt.title('Scheduled Departures')  
```

```python
import seaborn as sns
seaborn_plots=sns.catplot(x="City",y="Performed departures",kind='bar',data=airports)
seaborn_plots.set_xticklabels(rotation=90)
sns.set(rc={'figure.figsize':(20,50)})
```
